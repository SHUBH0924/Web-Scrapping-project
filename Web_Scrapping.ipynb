```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url = 'http://books.toscrape.com/'
response = requests.get(url)
response
```

```python
soup = bs(response.text)
```

```python
name = []
rating = []
price = []
link = []
```

# Name

```python
data = soup.find_all('h3')

for i in data:
    print(i.string)
    name.append(i.string)
```

```python
name
```

# Rating

```python
data_2 = soup.find_all('p', class_ = 'star-rating')

for i in data_2:
    print(i['class'][1])
    rating.append(i['class'][1])
```

```python
rating
```

# Price

```python
data_3 = soup.find_all('p', class_='price_color')

for i in data_3:
    prices = i.string.removeprefix('Â')
    price.append(prices)
```

```python
price
```

# Link

```python
link = []
for h3 in soup.find_all('h3'):
    for a in h3.find_all('a'):
        link.append(a.get('href').strip())
link
```

# automate

```python
for i in range(2,51):
  url = f'http://books.toscrape.com/catalogue/page-{i}.html'
  response = requests.get(url)
  soup = bs(response.text)
  data = soup.find_all('h3')
  for i in data:
   name.append(i.string)
   price.append(soup.find('p', class_='price_color').string.removeprefix('Â'))
   rating.append(soup.find('p', class_='star-rating').get('class')[1])
   link.append(i.find('a').get('href').strip())
```

```python
rating_mapping = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}

numerical_ratings = [rating_mapping[r] for r in rating]

print(numerical_ratings)
```

```python
price = [(p.replace('Â', '')) for p in price]
price
```

```python
for i in range (len(link)):
  link[i] = 'http://books.toscrape.com/' + link[i]
```

```python
des = []

for i in range(len(link)):
  url = link[i]
  response = requests.get(url)
  soup = bs(response.text)
  # (soup.find('p'))

  for i in soup.find_all('p'):
    des.append(i.get_text())

```

```python
data = {
    'Name': name,
    'Rating': numerical_ratings,  # Use the numerical ratings from the previous step
    'Price': price,
    'Link': link,
    # 'Description': des
}


# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('book_data.csv', index=False)
```
